```python

from pymilvus import MilvusClient

client = MilvusClient(url='http://127.0.0.1:19530')

```

```python
from pymilvus import  DataType,Function, FunctionType
from pymilvus.client.constants import ConsistencyLevel


# 创建数据库
databases = client.list_databases()
print("Available databases:", databases)


# 创建 Schema

schema = MilvusClient.create_schema(auto_id=True)
schema.add_field(field_name="id",
                  datatype=DataType.INT64,
                    is_primary=True,
                      auto_id=True,
                      max_length=32)
schema.add_field(field_name="header",
                  datatype=DataType.VARCHAR,
                    max_length=128)

# Define tokenizer parameters for text analysis


schema.add_field(field_name="context", 
                 datatype=DataType.VARCHAR,
                 max_length=65535,
                enable_match=True,  # Enable text matching
                enable_analyzer=True)

schema.add_field(field_name="vector", 
                 datatype=DataType.FLOAT_VECTOR,
                   dim=1024)




# 3.3. Prepare index parameters
index_params = client.prepare_index_params()

# 3.4. Add indexes
index_params.add_index(
    field_name="id",
    index_type=""
)

index_params.add_index(
    field_name="vector", 
    index_type="",
    metric_type=""
)

# Create the collection
client.create_collection(
    collection_name="hsbc_c",
    schema=schema,
    index_params=index_params
)

res = client.get_load_state(
    collection_name="hsbc_c"
)

print(res)
```
